```python
# 전체 항목에 대한 상관도를 구해보자
# 각 항목들의 상관도를 구해줌(-1~1)
#상관관계를 구했는데, 
# 0: 컬럼간에 전혀 상관이 없다.
# 1: 완벽히 상관이 있다.(양)
# -1: 완벽히 상관이 있다.(음)
# - + 0.3: 약간 상관이 있다.
# - + 0.5: 상관이 있는 것 같다.
# - + 0.7: 꽤 큰 상관이 있다.
raw_data.corr()
```
