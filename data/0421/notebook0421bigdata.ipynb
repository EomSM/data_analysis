```python
prac_pivot=practice2.pivot('날짜','운동', '양')
prac_pivot
#날짜: 인덱스 / 운동 : 열 종류/ 양: 테이블 내부의 값 
```

```python
prac_pivot.fillna(0, inplace=True)
#NaN 결측치를 삭제or 적합한 수치로 변환 (평균 or 0 or 특정 기본값)
# 변경을 원본에 반영하기 위해 inplace=True 
prac_pivot
```
