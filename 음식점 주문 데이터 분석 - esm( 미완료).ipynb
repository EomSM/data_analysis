```python
for i, item in enumerate(item_count.iteritems(), 0):
        print('TOP', i+1 ,': ', item)
#enumerate : 시리즈에서 인덱스도 반복해서 꺼내주고 내용도 꺼내주는. "열거하다",
    # jquery의 $('list').each에 해당
    #values만 빼지 않기 위해서 item_count.iteritems()라는 기능 부여 해야  
    #item에는 index도 tuple로 뽑아오기 함.
```

```python
item_count.tolist()
```

```python
for i, item in enumerate(item_count, 0):
        print('TOP', i+1 ,': ', item)    #values만 빼지는 경우기, index까지 필요. 
```

```python
for i, item in enumerate(item_count.tolist(), 0):
        print('TOP', i+1 ,': ', item)
        #series의 values만 리스트로 만듬 
```

```python
# enumerate 반복문: index랑, 반복한 item을 둘 다 꺼내고 싶을 때 쓰기 좋은 반복문
data100 =[100, 200, 300]
for i, item in enumerate(data100, 0): #0 = 시작값
    print(i, ":", item)
```

```python
for i, item in enumerate(item_count.iteritems(), 0):
        print('TOP', i+1 ,': ', item[0], item[1])
        #item tuple의 [0] 앞의 항목명 [1] 판매 개수
```

```python
for i, (pre, post) in enumerate(item_count.iteritems(), 0):
        print('TOP', i+1 ,': ', pre, post)
        #얘도 결국은 같은 결과
```
